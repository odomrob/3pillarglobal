```python
#NOW DEFINE A FUNCTION TO OPERATE ON EVERY INSTANCE OF A COLUMN

### DEFINE FIRST LETTER FUNCTION
## TAKE IN A STATE STRING
## RETURNS THE FIRST LETTER OF THAT STATE STRING

def grab_first_letter(state):
  return state[0]
```
